# Salvando Imagens redimensionadas

## 40x40 (q95)

```python
# Saving 40x40
from skimage.io import imread
from PIL import Image

path_coco = "/home/pesquisavcl/pesquisa/filipe/coco_images/originais/"
lista_imagens = sorted(glob(os.path.join(path_coco,"*jpeg")))

target_size = 40
output_folder = "/home/pesquisavcl/pesquisa/filipe/coco_images/tamanho40/originais/"
output_folder_jpeg = "/home/pesquisavcl/pesquisa/filipe/coco_images/tamanho40/jpeg/"
output_folder_webp = "/home/pesquisavcl/pesquisa/filipe/coco_images/tamanho40/webp/"

for im_path in lista_imagens:
    imagem_original = imread(im_path)
    
    name = im_path.split("/")[-1]
    print(name)
    
    im = Image.open(im_path).resize((target_size,target_size))
    rgb_im = im
    path_img_png = os.path.join(output_folder, name[:-4]+"png")
    rgb_im.save(path_img_png)
    
    #salvando para jpeg
    path_img_jpg = os.path.join(output_folder_jpeg, name)
    rgb_im.save(path_img_jpg,"JPEG", quality=95)
    
    #salvando para webp
    path_img_webp = os.path.join(output_folder_webp, name[:-4]+"webp")
    rgb_im.save(path_img_webp,"webp", quality=100)
    
```

```python
from skimage.metrics import mean_squared_error, peak_signal_noise_ratio, structural_similarity
from skimage.io import imread

path_coco_png = "/home/pesquisavcl/pesquisa/filipe/coco_images/tamanho40/originais"
# path_coco = "/Users/filipe/Downloads/coco_compress_paper/tamanho40_q50/jpeg"
path_coco = "/home/pesquisavcl/pesquisa/filipe/coco_images/tamanho40/webp"

lista_imagens_png = sorted(glob(os.path.join(path_coco_png,"*png")))
# lista_imagens = sorted(glob(os.path.join(path_coco,"*jpeg")))
lista_imagens = sorted(glob(os.path.join(path_coco,"*webp")))

print("Nome\tPNG\tJPEG\tCR")
lista_cr=[]
lista_psnr=[]
lista_mse =[]
lista_ssim=[]
for im in lista_imagens:
    nome = im.split("/")[-1]
    #path_img_original = os.path.join(path_coco_png,nome.replace("jpeg","png"))
    path_img_original = os.path.join(path_coco_png,nome.replace("webp","png"))
    tamanho_png = os.stat(path_img_original).st_size
    tamanho = os.stat(im).st_size
    cr = 100*(tamanho_png-tamanho)/tamanho_png
    lista_cr.append(cr)
    print(f"{im}\t{tamanho_png}\t{tamanho}\t{cr}")
    
    imagem_jpeg = imread(im)
    imagem_original = imread(path_img_original)
    psnr = peak_signal_noise_ratio(imagem_original, imagem_jpeg)
    lista_psnr.append(psnr)

    mse = mean_squared_error(imagem_original, imagem_jpeg)
    lista_mse.append(mse)
    
    ssim = structural_similarity(imagem_original, imagem_jpeg, win_size=3, multichannel=True)
    lista_ssim.append(ssim)
avg_cr = sum(lista_cr)/len(lista_cr)
avg_psnr = sum(lista_psnr)/len(lista_psnr)
avg_mse=sum(lista_mse)/len(lista_mse)
avg_ssim=sum(lista_ssim)/len(lista_ssim)
print(f"Media CR: {avg_cr}")
print(f"Media psnr: {avg_psnr}")
print(f"Media mse: {avg_mse}")
print(f"Media ssim: {avg_ssim}")
```
